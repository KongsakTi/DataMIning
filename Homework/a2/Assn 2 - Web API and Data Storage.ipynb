### 3.5  Kills/Death/Assist

Make a grouped horizonal barchart to display: total kills, total deaths, total assists for each hero on the same axis.

````python
# Fill code here
lst = []
for row in df.index:
    heros = df['players'][row]
    for idx, hero in enumerate(heros):
        try:
            hero_name = hero['hero_name']
            kills = hero["kills"]
            deaths = hero["deaths"]
            assists = hero["assists"]
                
            lst.append( {"hero_name": hero_name, "kills": kills, "deaths": deaths, "assists": assists} )
        except KeyError:
            pass
````

````python
# lst
````

````python
df_3 = pd.DataFrame(lst)
````

````python
df_3.groupby("hero_name").sum().sort_values("kills").plot(kind='barh',stacked=True, figsize=(20, 40))
````

### 3.6 Richest Heros

Find out the top 10 heros with highest Gold per Minute.

````python
# Fill code here
# Fill code here
lst = []
for row in df.index:
    heros = df['players'][row]
    for idx, hero in enumerate(heros):
        try:
            hero_name = hero['hero_name']
            gpm = hero["gold_per_min"]
                
            lst.append( {"hero_name": hero_name, "gold_per_min": gpm} )
        except KeyError:
            pass
````

````python
df_4 = pd.DataFrame(lst)
````

````python
df_4.groupby("hero_name").sum().sort_values("gold_per_min", ascending=False)[:10].plot(kind='barh',stacked=True, figsize=(10, 5))
````
